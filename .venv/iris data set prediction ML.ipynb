```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


```

LOADING DATASETS


```python
columns=['Id','Sepal Length','Sepal Width','Petal Length','Petal Width','Class']
df=pd.read_csv('Iris.csv')
df.head(50)
```

VISUALIZING DATASET

```python
df.describe()
```

```python

sns.pairplot(df,hue="Species")
```

SEPERATING INPUT AND OUTPUT COLUMNS


```python
#seperate features and targets
data=df.values
X=data[:,1:5]
Y=data[:,5]
print(X)
print(Y)
```

SPLITTING DATA (TO TRAIN AND TO TEST)

```python
#split the training and testing data
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
print(X_train)
print(Y_train)
print(X_test)
print(Y_test)
```

USING SVM

```python
from sklearn.svm import SVC
model_svc=SVC()
model_svc.fit(X_train,Y_train)
```

```python
#prediction
prediction1=model_svc.predict(X_test)
#calculating accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction1)*100)
for i in range(len(prediction1)):
    print(Y_test[i],prediction1[i]) #you can see wrong prediction
```

USING LOGISTIC RIGRESSION

```python
from sklearn.linear_model import LogisticRegression
model_LogisticRegression=LogisticRegression()
model_LogisticRegression.fit(X_train,Y_train)
```

```python
#prediction2
prediction2=model_LogisticRegression.predict(X_test)
#accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction2)*100)
```

USING DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
model_DTC=DecisionTreeClassifier()
model_DTC.fit(X_train,Y_train)
```

```python
#prediction
prediction3=model_DTC.predict(X_test)
#check accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction3)*100)
```

DETAILED CLASSIFICATION REPORT

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,prediction2))
```

PREDICTION FOR NEW SPECIES


```python
X_new=np.array([[3,2,1,0.2],[4.9,2.2,3.8,1.1],[6.7,3.1,5.6,2.4]])
prediction=model_svc.predict(X_new)
print("Prediction of new species:{}".format(prediction))
```
